```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define two lists
my_x = [ 0, 1, 1 ]
my_w = [ -10, 10, 5 ]
```

```python
# multiply two lists element-wise
def mul(a, b):
    return [a[i] * b[i] for i in range(len(a))]
```

```python
mul(my_x, my_w)
```

```python
my_b = -5
```

```python
def linear_neuron(x, weights, bias):
    return sum(mul(x, weights)) + bias 
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
for z in (-5, 0, 5):
    print(sigmoid(z))
```

```python
fig, ax = plt.subplots()
ax.axhline(c=".2"); ax.axvline(c=".2")
ax.grid(c='.1', ls='--')
z = np.linspace(-7.5, 7.5)
ax.plot(z, sigmoid(z))
plt.show()
```

```python
def neuron(x, weights, bias):
    return sigmoid(linear_neuron(x, weights, bias))
```

The affine function should be large for any nonzero inputs, but small for a zero input. So $$ b + w^T[0, 0] = b $$ should be small, but $$ b + w^T[1, 0] = b + w_1 $$ should be large. Thus $b$ should be small but $w_1$ and $w_2$ must be large relative to $b$.

```python
def or_neuron(x):
    """Return x1 OR x2."""
    return neuron(x, [10, 10], -5)
```

```python
my_x_collection = [[0,0], [0,1], [1,0], [1,1]]
```

```python
print("Checking OR neuron output")
for x in my_x_collection:
    print(f"{x}  {or_neuron(x):.1f}")
```

For AND, we need $$ b + w^T[1,1] = b + w_1 + w_2 $$ to be large, but $$ b + w^T[1,0] = b+w_1 $$ to be small. So we need $$ | b | > w_1 $$ and $$ |b | > w_2 $$ but $$ |b| < w_1 + w_2. $$

```python
def and_neuron(x):
    """Return x1 AND x2."""
    return neuron(x, [40, 40], -50)

print("Checking AND neuron output")
for x in my_x_collection:
    print(f"{x}  {and_neuron(x):.1f}")
```
